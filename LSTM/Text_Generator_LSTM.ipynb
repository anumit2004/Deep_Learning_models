```python
with open("/content/drive/MyDrive/dataset/shakespere_novel_dataset.txt" ,"r" , encoding="utf-8") as f:
  text = f.read()

text = text.lower()
print(text[:1000])
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokeniser = Tokenizer()
```

```python
tokeniser.fit_on_texts([text])
```

```python
vocab_size = len(tokeniser.word_index)
```

```python
vocab_size
```

```python
tokeniser.word_index
```

```python
clean_sentence = [sentence.strip() for sentence in text.split("\n") if sentence.strip()]
input_sequences = []
for sentence in clean_sentence:
    tokenised_sentence = tokeniser.texts_to_sequences([sentence])[0]
    for i in range(1, len(tokenised_sentence)):
      n_gram_sequence = tokenised_sentence[:i+1]
      input_sequences.append(n_gram_sequence)
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequence = pad_sequences(input_sequences , maxlen = max_len , padding = "pre")
```

```python
padded_input_sequence
```

```python
X = padded_input_sequence[:, :-1]
```

```python
Y  = padded_input_sequence[:,-1]
```

```python
X.shape
```

```python
Y.shape
```

```python
# Y will be used as integer labels for sparse_categorical_crossentropy, no need to one-hot encode
# from tensorflow.keras.utils import to_categorical
# Y = to_categorical(Y, num_classes = vocab_size+1)
```

```python
Y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input , Embedding , LSTM , Dense
```

```python
model = Sequential()
model.add(Input((max_len-1,)))
model.add(Embedding(input_dim = vocab_size+1 , output_dim = 100 ))
model.add(LSTM(1024))
model.add(Dense(vocab_size+1 , activation = "softmax"))
```

```python
model.compile(loss = "sparse_categorical_crossentropy" , optimizer="adam" , metrics = ["accuracy"])
```

```python
model.summary()
```

```python
model.fit(X,Y, epochs = 40 )
```

```python
import numpy as np

def generate_text(seed_text, next_words, model, tokenizer, max_sequence_len):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
        predicted_probs = model.predict(token_list, verbose=0)[0]
        predicted_word_index = np.argmax(predicted_probs)

        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted_word_index:
                output_word = word
                break
        seed_text += " " + output_word
    return seed_text
```

```python
seed_text = "first citizen: we are accounted poor citizens, the patricians good."
next_words = 10

generated_passage = generate_text(seed_text, next_words, model, tokeniser, max_len)
print(generated_passage)
```
